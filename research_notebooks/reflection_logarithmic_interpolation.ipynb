```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define original Dataset
x_values = np.array([315, 400, 500, 630, 800, 1000, 1250, 1600])
y_values = np.array([0.196625, 0.307625, 0.308625, 0.4655, 0.304, 0.344125, 0.381125, 0.481])

# Perform linear interpolation
model = np.polyfit(x_values, y_values, deg = 1)

# Generate new dataset
x_new_values = np.linspace(0,8000,81)
y_new_values = np.polyval(model, x_new_values)

# Plot Dataset
plt.figure()
plt.plot(x_values, y_values, 'x')
plt.title('Original Data')
plt.plot(x_new_values, y_new_values)
plt.title('Generated Data')
```

```python
# Transform Data in log-domain
log_x_values = np.log(x_values)
log_y_values = np.log(y_values)

# Perform linear interpolation in log-domain
log_model = np.polyfit(log_x_values, log_y_values, deg = 1)

# Generate new dataset
x_new_values = np.linspace(0.0001,16000,21)
y_new_log_values = np.polyval(log_model, np.log(x_new_values))

# Bring back to linear domain
y_new_values = np.exp(y_new_log_values)
y_new_values = np.clip(y_new_values, 0, 1)

# Plot Dataset
plt.figure()
plt.plot(x_values, y_values, 'x')
plt.title('Original Data')
plt.plot(x_new_values, y_new_values)
plt.title('Generated Data')
```
